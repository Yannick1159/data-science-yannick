```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans

# Simuleer klantdata
np.random.seed(42)
age = np.random.randint(18, 70, size=200)
spending = age * 15 + np.random.normal(0, 150, size=200)  # meer leeftijd = meer uitgaven (simulatie)

df = pd.DataFrame({'Leeftijd': age, 'JaarlijkseUitgaven': spending})

# KMeans clustering
kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(df[['Leeftijd', 'JaarlijkseUitgaven']])

# Plot resultaat
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Leeftijd', y='JaarlijkseUitgaven', hue='Cluster', palette='tab10', s=70)

# Plot de centroiden
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=250, marker='X', label='Centroiden')

plt.title("Klantensegmentatie met K-Means")
plt.xlabel("Leeftijd")
plt.ylabel("Jaarlijkse Uitgaven (â‚¬)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
